```python
from wordcloud import WordCloud

#      
from PIL import Image

#  
import matplotlib.pyplot as plt

#   
import numpy as np

#  
import pandas as pd
```

```python
#   
df = pd.read_csv('./fruit_vegetable.csv', encoding="utf-8")

#     
wc = df.set_index("title").to_dict()["count"]
```

```python
img = Image.open('./fruit_vegetable.png')
imgArray = np.array(img)    #     

worldCloud = WordCloud(
    font_path='malgun',
    width=400,
    height=400,
    max_font_size=100,
    background_color='white',
    mask=imgArray,
).generate_from_frequencies(wc)

plt.figure()
plt.imshow(worldCloud)
plt.axis('off')
```
