```python
#7.1 문서 인코딩

'''
문서 인코딩-> 애플리케이션(컴퓨터의 운영체제, 직접 만든 파이썬 코드...)이 그 문서를 읽는 방법을 지정

모든 문서는 근본적으로 0, 1로 인코드되어 있음.
인코딩 알고리즘은 그 위에서 글자 하나에 몇 비트인지, 이미지 파일이라면 각 픽셀에 몇 비트를 써서 색깔을 나타내는지 정의함.
그 위에 다시 PNG 파일처럼 압축이나 기타 공간을 절약하는 알고리즘이 있을 수 있다.

알맞은 라이브러리를 쓰면 파이썬은 어떤 형식의 정보라도 제대로 다룰 수 있다.
텍스트 파일과 비디오, 이미지 파일의 차이는 0과 1을 어떻게 해석하는가의 차이일 뿐이다.

텍스트, CSV, PDF, 워드 문서는 기본적으로 텍스트를 저장하는 파일 형식이다.
'''
```

```python
#7.3 CSV

from urllib.request import urlopen
from io import StringIO
import csv

url='http://pythonscraping.com/files/MontyPythonAlbums.csv'
data=urlopen(url).read().decode('ascii', 'ignore') #url을 받아서 읽고 decode한다.(ascii, ignore)
dataFile=StringIO(data)
csvReader=csv.reader(dataFile)
#csvReader가 반환하는 reader 객체는 순환체이며 파이썬 리스트 객체로 구성되어 있다.

for row in csvReader:   #csv.reader 메서드 대신 DictReader를 사용해보아
    print('The album"{}" nwas released in {}'.format(row[0], str(row[1])))
```

```python
from urllib.request import urlopen
from io import StringIO
import csv

url='http://pythonscraping.com/files/MontyPythonAlbums.csv'
data=urlopen(url).read().decode('ascii', 'ignore')
dataFile=StringIO(data)
dictReader=csv.DictReader(dataFile)
#csv.DictReader-> csv 파일의 각 행을 리스트 객체가 아니라 딕셔너리 객체러 반환
#필드 이름은 변수 dictReader.field에 저장되고 딕셔너리 객체의 키로도 저장된다.

print(dictReader.fieldnames)

for row in dictReader:
    print(row)
```
